#Q1 Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank=[]
name=[]
artist=[]
date=[]
view=[]
```

```python
ranks = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr')
for i in ranks:
    rk = i.text
    rank.append(rk)
names = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
for i in names:
    nm = i.text
    name.append(nm)
artists = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
for i in artists:
    art = i.text
    artist.append(art)
dates = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
for i in dates:
    dt = i.text
    date.append(dt)
views = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
for i in views:
    vw = i.text
    view.append(vw)
```

```python
df=pd.DataFrame({'Rank':rank,'NAME':name,'ARTIST':artist,'VIEWS':date,'DATE':view})
df
```

#Q6 Scrape the details of Highest selling novels.

C) Volumes sold
D) Publisher
E) Genre
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
driver=webdriver.Chrome()
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
Book=[]
Author =[]
Volume=[]
Publisher=[]
Genre=[]
```

```python
books= driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[2]')
for i in books:
    bk = i.text
    Book.append(bk)
auths= driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[3]')
for i in auths :
    aut = i.text
    Author.append(aut)
vols= driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[4]')
for i in vols:
    vl = i.text
    Volume.append(vl)
PUBS = driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[5]')
for i in PUBS :
    pb = i.text
    Publisher.append(pb)
gens= driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[6]')
for i in gens:
    gr = i.text
    Genre.append(gr)
```

```python
len(Author)
```

B) Year span
C) Genre
D) Run time

F) Votes 

```python
driver=webdriver.Chrome()
driver.get(' https://www.imdb.com/list/ls512407256/')
```

```python
Name=[]
Year=[]
Genre=[]
Runtime=[]
Ratings=[]
Votes =[]
```

```python
gens= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/p/span[5]')
for i in gens:
    gr = i.text
    Genre.append(gr)
    
Names= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/h3/a')
for i in Names:
    nm = i.text
    Name.append(nm)
    
Years= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/h3/span[2]')
for i in Years:
    yr = i.text
    Year.append(yr)

Runtimes= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/p/span[3]')
for i in Runtimes :
    rt= i.text
    Runtime.append(rt)
    
Ratingss= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/div/div/span[2]')
for i in Ratingss:
    rg = i.text
    Ratings.append(rg)
    
Votess= driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]/div[2]/p[4]/span[2]')
for i in Votess:
    vt = i.text
    Votes.append(vt)
    
```

```python
df=pd.DataFrame({'Name':Name,'Year':Year,'Genre':Genre,'Runtime':Runtime,'Ratings':Ratings,'Votes':Votes})
df
```
