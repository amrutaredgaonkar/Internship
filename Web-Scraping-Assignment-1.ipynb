```python
#1) Write a python program to display all the header tags from wikipedia.org and make data frame.

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
first_title=soup.find('span',class_="mw-headline")
first_title
```

```python
first_title.text
```

```python
titles=[]
for i in soup.find_all('span',class_="mw-headline"):
    titles.append(i.text)
titles
```

```python
df=pd.DataFrame({'TITLE':titles})
df
```

```python
#Q2 Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
```

```python
page=requests.get('https://presidentofindia.nic.in/')
page
soup=BeautifulSoup(page.content)
soup
```

```python
president=soup.find('h3',class_="wgl-services_title")
president
```

```python
president.text
```

```python
titles=[]
for i in soup.find_all('h3',class_="wgl-services_title"):
    titles.append(i.text)
titles
```

```python
time=soup.find('div',class_="wgl-services_text")
time
```

```python
time.text
```

```python
times=[]
for i in soup.find_all('div',class_="wgl-services_text"):
    times.append(i.text)
times
```

```python
df=pd.DataFrame({'TITLE':titles,'PERIOD':times})
df
def extract_dates(text):
    pattern = r'\d{1,2} (?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}'
    matches = re.findall(pattern, text)
    return matches   

df['DATES'] = df['PERIOD'].apply(extract_dates)
df.drop(columns=['PERIOD'], inplace=True)
print(df)
```
